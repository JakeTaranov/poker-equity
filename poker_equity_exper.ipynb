```python
def check_high_card(holding, board):
    combined = holding + board
    ranks = [rank for rank, _ in combined]
    return sum(sorted(ranks, reverse=True)[:5])


def t_check_pair(holding, board):
    combined = holding + board
    cur_max_pair = 0
    rank_counts = {}

    for rank, _ in combined:
        rank_counts[rank] = rank_counts.get(rank, 0) + 1

    for rank, count in rank_counts.items():
        if count == 2:
            cur_max_pair = max(cur_max_pair, rank)
            
    return cur_max_pair


def check_pair(holding, board):    
    board_pair = t_check_pair([], board)
    board_hand_pair = t_check_pair(holding, board)
    
    
    # logic is similar to two pair, read comments if confused
    if board_pair == board_hand_pair:
        return board_pair   
    else:
        return board_hand_pair
    


def t_two_pair(holding, board):
    combined = holding + board
    rank_counts = {}
    cur_max_two_pairs = 0
    
    for rank, _ in combined:
        rank_counts[rank] = rank_counts.get(rank, 0) + 1
        
    pairs = [rank for rank, count in rank_counts.items() if count == 2]
    
    pairs = sorted(pairs, reverse=True)
    if len(pairs) >= 2:
        cur_max_two_pairs = [pairs[0], pairs[1]]
    
    return cur_max_two_pairs

def check_two_pair(holding, board):
    
    board_two_pair = t_two_pair([], board)
    board_and_hand_two_pair = t_two_pair(holding, board)
    
    # if the pair top pair on the board is the same as the top pair with holdings then return the top pair (either)
    if board_two_pair == board_and_hand_two_pair:
        return board_two_pair
    else:
    # if the pair on the board is not the same as the pair with the players hand, then this means
    # the pair with the players hand must be higher since we return the max 
        return board_and_hand_two_pair
    


    

def find_unique_in_three_of_a_kind(board):
    count = {}
    for rank, _ in board:
        count[rank] = count.get(rank, 0) + 1
    
    unique = [rank for rank, count in count.items() if count == 1]
    return unique
    


def check_three_of_a_kind(holding, board):
    combined = holding + board
    cur_max_trips = 0
    rank_counts = {}
    
    # if there are three of a kind on the board then we need to get the max from the players cards and cards on the board
    if holding != [(None, None)]:
        board_has_three_of_a_kind = check_three_of_a_kind([(None, None)], board)
        
        if board_has_three_of_a_kind:
            holding_ranks = [rank for rank, _ in holding]
            unique_in_board = find_unique_in_three_of_a_kind(board)
            possible_cards = holding_ranks + unique_in_board
            return max(possible_cards)
        
        
    for rank, _ in combined:
        rank_counts[rank] = rank_counts.get(rank, 0) + 1
    
    for rank, count in rank_counts.items():
        if count == 3:
            cur_max_trips = max(cur_max_trips, rank)
    
    return cur_max_trips
    
    
    
def check_straight(holding, board):
    combined = holding + board
    rank_set = sorted(set(rank for rank, _ in combined))
    straight_score = 0
    
    for i in range(len(rank_set) - 4):
        #checks to see if there is any straight
        if rank_set[i] == rank_set[i + 1] - 1 == rank_set[i + 2] - 2 == rank_set[i + 3] - 3 == rank_set[i + 4] - 4:
            straight_score = max(straight_score ,rank_set[i] + rank_set[i+1] + rank_set[i+2] + rank_set[i+3] + rank_set[i+4])
    
    #checks to see if there is an ace -> 5 straight
    if rank_set[-1] == 14 and rank_set[0] == 2 and rank_set[1] == 3 and rank_set[2] == 4 and rank_set[3] == 5:
        straight_score = max(straight_score ,rank_set[i] + rank_set[i+1] + rank_set[i+2] + rank_set[i+3] + rank_set[i+4])
    
    return straight_score

def check_flush(holding, board):
    combined = holding + board
    max_rank_flush = 0
    suit_counts = {}
    
    for rank, suit in combined:
        if suit not in suit_counts:
            suit_counts[suit] = [rank]
        else:
            suit_counts[suit].append(rank)
            
    for suit, ranks in suit_counts.items():
        if len(ranks) >= 5:
            max_rank_flush = max(max_rank_flush, sum(sorted(ranks, reverse=True)[:5]))
            
    return max_rank_flush

def check_full_house(holding, board):
    combined = holding + board
    
    rank_counts = {}
    
    three_of_a_kind = False
    two_of_a_kind = False
    max_full_three = 0
    max_full_two = 0


    for rank, _ in combined:
        rank_counts[rank] = rank_counts.get(rank, 0) + 1
        
    for rank, count in rank_counts.items():
        if count == 3:
            three_of_a_kind = True
            max_full_three = max(max_full_three, rank)
        elif count == 2:
            two_of_a_kind = True
            max_full_two = max(max_full_two, rank)
    
    if three_of_a_kind and two_of_a_kind:
        return max_full_three*3 + max_full_two*2
    return 0


def find_unique_in_four_of_a_kind(board):
    count = {}
    
    for rank, _ in board:
        count[rank] = count.get(rank, 0) + 1
        
    unique = [rank for rank, count in count.items() if count == 1]
    return unique

def check_four_of_a_kind(holding, board):
    combined = holding + board
    cur_max_quads = 0
    rank_counts = {}
    
    # if there is four of a kind on the board, then the highest card from holding and reamining card on board plays
    
    if holding != [(None, None)]:
        board_has_four_of_a_kind = check_four_of_a_kind([(None, None)], board)

        if board_has_four_of_a_kind:
            holding_ranks = [rank for rank, _ in holding]
            unique_in_board = find_unique_in_four_of_a_kind(board)
            possible_cards = holding_ranks + unique_in_board
            return max(possible_cards)
            
        
            
    for rank, _ in combined:
        rank_counts[rank] = rank_counts.get(rank, 0) + 1
        
    
    for rank, count in rank_counts.items():
        if count >= 4:
            cur_max_quads = max(cur_max_quads, rank)
    
    return cur_max_quads
    

def check_straight_flush(holding, board):
    
    has_flush = check_flush(holding, board)
    has_straight = check_straight(holding, board)
    max_straight_flush = 0
    
    if has_flush and has_straight:
        #returns the max value of the straight
        max_straight_flush = has_straight
    
    return max_straight_flush
        

def check_royal_flush(holding, board):
    combined = holding + board
    
    if check_straight(holding, board) and check_flush(holding, board):
        straight_flush_ranks = [rank for rank, _ in combined if rank in {10, 11, 12, 13, 14}]
        
        if set(straight_flush_ranks) == {10, 11, 12, 13, 14}:
            return 1
        
    return 0
```

```python
cards = [
    { "suit": "spade", "value": "2", "rank": 2, "isPlayable": True },
    { "suit": "spade", "value": "3", "rank": 3, "isPlayable": True },
    { "suit": "spade", "value": "4", "rank": 4, "isPlayable": True },
    { "suit": "spade", "value": "5", "rank": 5, "isPlayable": True },
    { "suit": "spade", "value": "6", "rank": 6, "isPlayable": True },
    { "suit": "spade", "value": "7", "rank": 7, "isPlayable": True },
    { "suit": "spade", "value": "8", "rank": 8, "isPlayable": True },
    { "suit": "spade", "value": "9", "rank": 9, "isPlayable": True },
    { "suit": "spade", "value": "10", "rank": 10, "isPlayable": True },
    { "suit": "spade", "value": "J", "rank": 11, "isPlayable": True },
    { "suit": "spade", "value": "Q", "rank": 12, "isPlayable": True },
    { "suit": "spade", "value": "K", "rank": 13, "isPlayable": True },
    { "suit": "spade", "value": "A", "rank": 14, "isPlayable": True },
    { "suit": "heart", "value": "2", "rank": 2, "isPlayable": True },
    { "suit": "heart", "value": "3", "rank": 3, "isPlayable": True },
    { "suit": "heart", "value": "4", "rank": 4, "isPlayable": True },
    { "suit": "heart", "value": "5", "rank": 5, "isPlayable": True },
    { "suit": "heart", "value": "6", "rank": 6, "isPlayable": True },
    { "suit": "heart", "value": "7", "rank": 7, "isPlayable": True },
    { "suit": "heart", "value": "8", "rank": 8, "isPlayable": True },
    { "suit": "heart", "value": "9", "rank": 9, "isPlayable": True },
    { "suit": "heart", "value": "10", "rank": 10, "isPlayable": True },
    { "suit": "heart", "value": "J", "rank": 11, "isPlayable": True },
    { "suit": "heart", "value": "Q", "rank": 12, "isPlayable": True },
    { "suit": "heart", "value": "K", "rank": 13, "isPlayable": True },
    { "suit": "heart", "value": "A", "rank": 14, "isPlayable": True },
    { "suit": "diamond", "value": "2", "rank": 2, "isPlayable": True },
    { "suit": "diamond", "value": "3", "rank": 3, "isPlayable": True },
    { "suit": "diamond", "value": "4", "rank": 4, "isPlayable": True },
    { "suit": "diamond", "value": "5", "rank": 5, "isPlayable": True },
    { "suit": "diamond", "value": "6", "rank": 6, "isPlayable": True },
    { "suit": "diamond", "value": "7", "rank": 7, "isPlayable": True },
    { "suit": "diamond", "value": "8", "rank": 8, "isPlayable": True },
    { "suit": "diamond", "value": "9", "rank": 9, "isPlayable": True },
    { "suit": "diamond", "value": "10", "rank": 10, "isPlayable": True },
    { "suit": "diamond", "value": "J", "rank": 11, "isPlayable": True },
    { "suit": "diamond", "value": "Q", "rank": 12, "isPlayable": True },
    { "suit": "diamond", "value": "K", "rank": 13, "isPlayable": True },
    { "suit": "diamond", "value": "A", "rank": 14, "isPlayable": True },
    { "suit": "club", "value": "2", "rank": 2, "isPlayable": True },
    { "suit": "club", "value": "3", "rank": 3, "isPlayable": True },
    { "suit": "club", "value": "4", "rank": 4, "isPlayable": True },
    { "suit": "club", "value": "5", "rank": 5, "isPlayable": True },
    { "suit": "club", "value": "6", "rank": 6, "isPlayable": True },
    { "suit": "club", "value": "7", "rank": 7, "isPlayable": True },
    { "suit": "club", "value": "8", "rank": 8, "isPlayable": True },
    { "suit": "club", "value": "9", "rank": 9, "isPlayable": True },
    { "suit": "club", "value": "10", "rank": 10, "isPlayable": True },
    { "suit": "club", "value": "J", "rank": 11, "isPlayable": True },
    { "suit": "club", "value": "Q", "rank": 12, "isPlayable": True },
    { "suit": "club", "value": "K", "rank": 13, "isPlayable": True },
    { "suit": "club", "value": "A", "rank": 14, "isPlayable": True },
];



def shuffle_deck():
    random.shuffle(cards)
    


def check_winner(h1, h2):
    if h1 > h2:
        return 1
    elif h2 > h1:
        return 2
    elif h1 == h2 and h1 != 0 and h2 != 0:
        return 3
    else:
        return 0
        

def check_two_pair_winner(h1,h2, hand_1, hand_2, board):
    
    
    # print("H1", h1)
    # print("hand1", hand_1)
    # print("H2", h2)    
    # print('hand2', hand_2)
    # print(board)
    # print('\n\n')
    if h1 == 0 and h2 == 0:
        # no two pairs found
        #print("NO TWO PAIRS FOUND")
        return 0
    if h2 == 0 and h1:
        return 1
    if h1 == 0 and h2:
        return 2

    

 #   print(h1, h2)

    if h1[0] > h2[0]:
        #print("H1 wins")
        return 1
    if h2[0] > h1[0]:
        #print("h2 wins")
        return 2
    else:
        # if the top pairs are equal then we check the second pair    
        if h1[1] == h2[1]:
            # if these are equal then the high card from either hand wins
            hand_1_max = max([rank for rank, _ in hand_1])
            hand_2_max = max([rank for rank, _ in hand_2])
            
            board_rank_count = {}
            for rank, _ in board:
                board_rank_count[rank] = board_rank_count.get(rank, 0) + 1
            unique_in_board = max([rank for rank, count in board_rank_count.items() if count == 1])
            
            # If the last card in the board is higher than the max in hand 1 and hand 2, then its a tie
            if unique_in_board > hand_1_max and unique_in_board > hand_2_max:
                # split
                return 3    
            
            if hand_1_max > hand_2_max:
                # hand 1 wins from high card
                return 1
            elif hand_2_max > hand_1_max:
                # hand 2 wins from high card
                return 2
            else:
                # equal high cards and pairs, split pot
               # print("ASDDDDDDDDDDDDDDDDDDDDDDDDDD")
                return 3
            #print("TIE")
        elif h1[1] > h2[1]:
           # print("H1 wins from second pair")
           return 1
        else:
            #print("H2 WINS from second pair")
            return 2
            
def check_pair_winner(h1, h2, hand_1, hand_2):
    #print(h1, h2)
    if h2 == 0 and h1:
        return 1
    if h1 == 0 and h2:
        return 2
    if h1 == 0 and h2 == 0:
        # no pairs found
        #print("NO TWO PAIRS FOUND")
        return 0 

    if h1 > h2:
        #h1 wins
        return 1
    elif h2 > h1:
        #h2 wins
        return 2
    else:
        # they are equal so we check kickers
        hand_1_max = max([rank for rank, _ in hand_1])
        hand_2_max = max([rank for rank, _ in hand_2])
        if hand_1_max > hand_2_max:
            return 1
        if hand_2_max > hand_1_max:
            return 2
        else:
            return 3
        
def simulate_hand(hand_1, hand_2, board):


    h1s = check_royal_flush(hand_1, board)
    h2s = check_royal_flush(hand_2, board)
    winner = check_winner(h1s, h2s)
    
    if winner == 1:
      #  print("HAND 1 WON BY ROYAL FLUSH")
        return 1, "r"
    if winner == 2:
     #   print("HAND 2 WON BY ROYAL FLUSH")
        return 2, "r"
    if winner == 3:
        # print("TIE ROYAL FLUSH")
        # print(board)
        # print('\n', hand_1)
        # print('\n', hand_2)
        return 3, "r"
    
    h1s = check_straight_flush(hand_1, board)
    h2s = check_straight_flush(hand_2, board)
    winner = check_winner(h1s, h2s)
    
    if winner == 1:
     #   print("HAND 1 WON BY STRAIGTH FLUSH")
        return 1, "sf"
    if winner == 2:
      #  print("HAND 2 WON BY STRAIGTH FLUSH")
        return 2, "sf"
    if winner == 3:
        # print("TIE STRAIGHT FLUSH")
        # print(board)
        # print('\n', hand_1)
        # print('\n', hand_2)
        return 3, "sf"
    h1s = check_four_of_a_kind(hand_1, board)
    h2s = check_four_of_a_kind(hand_2, board)
    winner = check_winner(h1s, h2s)
    
    if winner == 1:
     #   print("HAND 1 WON BY FOUR")
        return 1, "4"
    if winner == 2:
       # print("HAND 2 WON BY FOUR")
        return 2, "4"
    if winner == 3:
        # print("TIE FOUR")
        # print(board)

        return 3, "4"
    
    h1s = check_full_house(hand_1, board)
    h2s = check_full_house(hand_2, board)
    winner = check_winner(h1s, h2s)
    
    if winner == 1:
     #   print("HAND 1 WON BY FH")
        return 1, "fh"
    if winner == 2:
      #  print("HAND 2 WON BY FH")
        return 2, "fh"
    if winner == 3:
        # print("TIE FH")
        # print(board)

        return 3, "fh"
    
    h1s = check_flush(hand_1, board)
    h2s = check_flush(hand_2, board)
    winner = check_winner(h1s, h2s)
    
    if winner == 1:
       # print("HAND 1 WON BY  FLUSH")
        return 1, "f"
    if winner == 2:
      #  print("HAND 2 WON BY  FLUSH")
        return 2, "f"
    if winner == 3:
        # print("TIE FLUSH")
        # print(board)

        return 3, "f"
    
    h1s = check_straight(hand_1, board)
    h2s = check_straight(hand_2, board)
    winner = check_winner(h1s, h2s)
    
    if winner == 1:
       # print("HAND 1 WON BY STRAIGTH ")
        return 1, "s"
    if winner == 2:
      #  print("HAND 2 WON BY STRAIGTH ")
        return 2, "s"
    if winner == 3:
        # print("TIE STRAIGTH")
        # print(board)

        return 3, "s"
    
    h1s = check_three_of_a_kind(hand_1, board)
    h2s = check_three_of_a_kind(hand_2, board)
    winner = check_winner(h1s, h2s)
    
    if winner == 1:
      #  print("HAND 1 WON BY THREE")
        return 1, "3"
    if winner == 2:
      #  print("HAND 2 WON BY THREE")
        return 2, "3"
    if winner == 3:
        # print("TIE THREE")
        # print(board)

        return 3, "3"
    
    h1s = check_two_pair(hand_1, board)
    h2s = check_two_pair(hand_2, board)
    winner = check_two_pair_winner(h1s, h2s, hand_1, hand_2, board)
    
    if winner == 1:
       # print("HAND 1 WON BY TWO PAIR")
        return 1, "2p"
    if winner == 2:
      #  print("HAND 2 WON BY TWO PAIR")
        return 2, "2p"    
    if winner == 3:
        # print("TIE TWO PAIR")
        # print(board)

        return 3, "2p"
    
    
    h1s = check_pair(hand_1, board)
    h2s = check_pair(hand_2, board)
    winner = check_pair_winner(h1s, h2s, hand_1, hand_2)
    
    if winner == 1:
       # print("HAND 1 WON BY PAIR")
        return 1, "p"
    if winner == 2:
       # print("HAND 2 WON BY PAIR")
        return 2, "p"
    if winner == 3:
       # print("TIE PAIR")
#         print(board)
#         print('\n', hand_1)
#         print('\n', hand_2)
        return 3, "p"
    
    h1s = check_high_card(hand_1, board)
    h2s = check_high_card(hand_2, board)
    winner = check_winner(h1s, h2s)
    #if h1s == h2s:
      #  print("EQ")
    if winner == 1:
       # print("HAND 1 WON BY HIGH")
        return 1, "h"
    if winner == 2:
      #  print("HAND 2 WON BY HIGH")
        return 2, "h" 
    if winner == 3:
        # print("TIE HIGH")
        # print(board)
        return 3, "h"
    

def make_board():
    random_idx = random.sample(range(48), 5)
    board = []
    for idx in random_idx:
        board.append((cards[idx]["rank"], cards[idx]["suit"]))
    return board

def run_n_hands(n):
    h1c1 = cards.pop(random.randint(0,51))
    h1c2 = cards.pop(random.randint(0,50))
    hand1 = [(h1c1["rank"], h1c1["suit"]), (h1c2["rank"], h1c2["suit"])]
    
    #hand1 = [(9, 'diamond'), (9, 'heart')]

    h2c1 = cards.pop(random.randint(0,49))
    h2c2 = cards.pop(random.randint(0,48))
    hand2 = [(h2c1["rank"], h2c1["suit"]), (h2c2["rank"], h2c2["suit"])]
    
   # hand2 = [(10, 'diamond'), (12, 'spade')]
    
    
    hand1_wins, hand2_wins, split = 0, 0, 0
    
    
    print(hand1)
    print(hand2)
    
    split_info = {}
    h1_info = {}
    h2_info = {}
    
    
    for i in range(n):
        shuffle_deck()
        board = make_board()
     #   board = [(7, 'spade'), (8, 'spade'), (9, 'spade'), (10, 'spade'), (11, 'spade')]
        res, how = simulate_hand(hand1, hand2, board)
       # print(board)
#         print(board)
        if res == 1:
            hand1_wins += 1
            h1_info[how] = h1_info.get(how, 0) + 1
            # if how == "2p":
            #     print(board)

#             print("H1 wins with", hand1)
        if res == 2:
#             print("H2 wins with", hand2)
            hand2_wins += 1
            h2_info[how] = h2_info.get(how, 0) + 1

        if res == 3:
            split += 1
            split_info[how] = split_info.get(how, 0) + 1
            # if how == "fh":
            #      print(board)
#         print('\n\n\n')
            
    return hand1_wins, hand2_wins, split, split_info, h1_info, h2_info
        

n = 100000
h1_w, h2_w, split, split_info, h1_info, h2_info = run_n_hands(n)
print(h1_w)
print(h2_w)
#draws = n-(h1_w + h2_w)
print(split)

print("Hand 1 wins", h1_w/n * 100)
print("Hand 2 wins", h2_w/n * 100)
print("Split ", split/n * 100)

print('\n')

for how, data in split_info.items():
    print(how, data)
    
print('\n\n H1')

for how, data in h1_info.items():
    print(how, data)
    
print('\n\n H2')

for how, data in h2_info.items():
    print(how, data)
#print(split_info['fh'])


# print(" Hand 1 wins", h1_w, "\n Hand 2 wins", h2_w, "\n Draws", n - (h1_w + h2_w))    
    
    
```
